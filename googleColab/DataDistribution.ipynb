# Data Distribution

# Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Faking a data distribution of rolling a dice

```python
x = np.random.normal(5, 1, 1000) #mean = 5, standard deviation = 1, number of samples= 1000

#Graphical representation of normal distribution
plt.hist(x)
plt.show()
```

# Uniform Distribution

```python
x = np.random.uniform(1, 6, 10000)
plt.hist(x)
plt.show()
```

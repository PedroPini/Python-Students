# Exploratory Data Analysis

**Exploratory Data Analysis (EDA)** is the process of examining and understanding your data before applying any machine learning models.

Think of it like **getting to know your dataset** — spotting patterns, checking for errors, and figuring out what’s important.

# Import Packages

```python
import pandas as pd #for data analysis
import seaborn as sns # for graphical visualisation
import matplotlib.pyplot as plt # for plotting

from sklearn.datasets import load_iris
```

# Load Iris

```python
iris = load_iris()
print(iris)
```

# Create the dataframe

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

print(df)
```

# Summary of the Dataset

| Term                | What it means (in plain English)                                            |
| ------------------- | --------------------------------------------------------------------------- |
| **count**           | Number of values (rows) in the column. Helps spot missing values.           |
| **mean**            | The _average_ — sum of all values divided by how many values there are.     |
| **std**             | Standard deviation — tells you how _spread out_ the values are.             |
| **min**             | The _smallest_ value in the column.                                         |
| **25% / 50% / 75%** | These are _percentiles_. For example, 50% is the **median** (middle value). |
| **max**             | The _largest_ value in the column.                                          |

```python
df.describe()
```

# Checking for null values

```python
#find null values and remove it so that it doesn't affect the data
df.isnull().sum()

#drop or remove null values
# df = df.dropna()
```

# Graphical Analysis

```python
# Create histograms for each of the data
df.hist(bins=10, figsize=(10,10)) # bins=10 means split the range of values into 10 intervals | figsize=(10, 10) size of the entire figure (the full plot window)
plt.show()
```

# Correlation Between Variables


Positive between 0 to +1
Negative between 0 to -1

```python
corr_matrix = df.corr()
corr_matrix
```

# Correlation Matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# Pairplot

```python
sns.pairplot(df)
plt.show()
```

## Use the library

```python
import matplotlib.pyplot as plt
```

# Basic Line Pot

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

plt.plot(x, y)
plt.show()
```

# Titles and Labels

```python
plt.title('My first Plot')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

plt.plot(x, y)
plt.show()
```

# Multiple Lines

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

plt.plot(x, y)

plt.plot(x, y, label="Line A")
plt.plot(x, [i + 2 for i in y], label="Line B")
plt.legend()
```

# Bar Chart

```python
categories = ['A', 'B', 'C']
values = [4, 7, 2]
plt.bar(categories, values)
plt.show()
```

# Scatter Plot

```python
# Sample data: Height vs Weight of a few individuals
heights = [150, 160, 165, 170, 175, 180, 185]
weights = [50, 55, 60, 65, 70, 75, 80]

# Create scatter plot
plt.scatter(heights, weights)

# Add context with labels and title
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

plt.show()
```

# Save your plot

```python
# Sample data: Height vs Weight of a few individuals
heights = [150, 160, 165, 170, 175, 180, 185]
weights = [50, 55, 60, 65, 70, 75, 80]

# Create scatter plot
plt.scatter(heights, weights)

# Add context with labels and title
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.savefig("myplot.png") ## Save before plt.show()
```
